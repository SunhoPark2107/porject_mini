```python
import pandas as pd
```

```python
hotel_df = pd.read_excel('hotel_bookings.xlsx')
```

```python
hotel_df.head()
```

```python
hotel_df.info()
```

```python
# 결측치 확인 - children, country, agent, company 결측치 존재.
hotel_df.isnull().sum()
```

```python
# 특성 삭제 - arrival_date_week_number, hotel, distribution_channel, company데이터 셋 삭제.
hotel_df = hotel_df.drop(['arrival_date_week_number', 'hotel', 'distribution_channel', 'company'], axis = 1)
```

```python
# 특성 대치 - children, agent, country
hotel_df['children'] = hotel_df['children'].fillna(0)
hotel_df['agent'] = hotel_df['agent'].fillna(-1)
hotel_df['country'] = hotel_df['country'].fillna('undefined')
```
